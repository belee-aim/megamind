```python
from llama_cloud_services import LlamaParse

# from megamind.utils.config import settings


class Parser:
    def __init__(self):
        self.parser = LlamaParse(
            api_key="llx-3feYJfEkK6ypqi4DyL6P7AkRgZf2kbzFCp1yeeaK1dw1qFKN",
            num_workers=4,
            result_type="markdown",
            verbose=True,
        )

    def parse(self, content: str, mime_type: str) -> dict:
        """
        Parse the content based on its MIME type.
        """
        return self.parser.parse(content, mime_type=mime_type)

    def parse_file(self, file_path: str) -> dict:
        """
        Parse a file and return its content.
        """
        result = self.parser.parse(file_path)
        markdown_nodes = result.get_markdown_documents(split_by_page=True)
        langchain_documents = [markdown_node.to_langchain_format() for markdown_node in markdown_nodes]

        return langchain_documents


parser = Parser()

markdown_nodes = parser.parse_file("../docs/+IR_Dec24_to_publish.pdf")
```
